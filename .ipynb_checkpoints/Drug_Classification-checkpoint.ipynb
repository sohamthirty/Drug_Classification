## Encoding Categorical Variables

```python
df.dtypes
```

```python
obj_df = df[['Sex','BP','Cholesterol','Drug']]
```

```python
obj_df.head()
```

### Encoding independent variable

```python
for i in obj_df.columns:
    print(i)
    print(obj_df[i].value_counts())
    print("-----------------------")
```

```python
#One Hot Encoding using pd.get_dummies
pd.get_dummies(obj_df, columns=['Sex']).head()
```

```python
#One Hot Encoding using sklearn
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()

enc_result = enc.fit_transform(df[['Sex']])

pd.DataFrame(enc_result.toarray(), columns = enc.categories_).head()
```

```python
enc_result
```

```python
#Ordinal using sklearn
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
df['BP_enc'] = enc.fit_transform(obj_df[['BP']])
df['Cholesterol_enc'] = enc.fit_transform(obj_df[['Cholesterol']])

df
```

### Encoding dependent variable

```python
#Ordinal using sklearn
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
df['Drug_enc'] = enc.fit_transform(obj_df[['Drug']])
```

## Splitting the dataset into the Training set and Test set

```python
d = df.drop(columns=['Sex','BP','Cholesterol','Drug'])
```

```python
d
```

```python
X = d.iloc[:, :-1].values
y = d.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
pd.reset_option('display')
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# To find the correlation using pearson method
df.corr(method ='pearson')
```

```python
# Heat Map
import seaborn as sns; sns.set_theme()

ax = sns.heatmap(df.corr(method ='pearson'),linewidths=.5, center=0)
```

## Data Modelling

```python
results = {}
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(X_train,y_train)
```

```python
print(classifier.predict([[23,7.298,1.0,0.0]]))
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
results["Decision Tree"] = accuracy_score(y_test, y_pred)
```

```python
# Print Text Representation
from sklearn import tree
text_representation = tree.export_text(classifier)
print(text_representation)
```

```python
# Plot Tree with plot_tree
plt.figure(figsize=(20,20))
tree.plot_tree(classifier,filled=True)
plt.title("plot_tree for Decision tree classifier")
plt.show()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
results["Random Forest"] = accuracy_score(y_test, y_pred)
```

### SVM

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
results["SVM"] = accuracy_score(y_test, y_pred)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
results["Logistic Regression"] = accuracy_score(y_test, y_pred)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
results["KNN"] = accuracy_score(y_test, y_pred)
```

### Kernel SVM

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
results["Kernel SVM"] = accuracy_score(y_test, y_pred)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
results["Naive Bayes"] = accuracy_score(y_test, y_pred)
```

## Results

```python
result_df = pd.DataFrame.from_dict(results ,orient = "index",columns=["Score"])
```

```python
plt.figsize=(20,5)

sns.barplot(x = result_df.index,y = result_df.Score)

plt.xticks(rotation=60)

plt.show()
```

## Feature Importance

```python
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
model = classifier.fit(X_train, y_train)
importance_value=np.round((model.feature_importances_)*100,2)
```

```python
importance_value
```

```python
importance_features=[(feature,importance) for feature,importance in zip(df.columns,importance_value)]
importance_features=sorted(importance_features,key=lambda x: x[1],reverse=True) 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in importance_features];
```

```python
type(importance_features)
```

```python
importance_df = pd.DataFrame(importance_features, columns=['Variable', 'Importance_Score'])
```

```python
plt.figsize=(20,5)

sns.barplot(x = importance_df.Variable,y = importance_df .Importance_Score)

plt.xticks(rotation=60)

plt.show()
```
